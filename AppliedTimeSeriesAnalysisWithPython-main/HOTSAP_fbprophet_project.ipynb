# Forecast Bus Ridership 

```python
import numpy as np
import pandas as pd
from itertools import product
from fbprophet import Prophet
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from fbprophet.plot import plot_cross_validation_metric
from fbprophet.diagnostics import cross_validation, performance_metrics

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10, 7.5)
```

## Import data 

```python
df = pd.read_csv('portland-oregon-average-monthly-.csv')
df.head()
```

```python
df.columns = ['ds', 'y']
df.head()
```

```python
df.tail()
```

```python
df.describe().transpose()
```

```python
fig, ax = plt.subplots()

locator = mdates.AutoDateLocator()

ax.plot(df.ds, df.y)
ax.xaxis.set_major_locator(locator)

ax.set_xlabel('Date')
ax.set_ylabel('Bus ridership')

ax.set_title('Monthly Bus Ridership in Portland, Oregon')

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

## Hyperparameter tuning 

```python
param_grid = {
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0]
}

params = [dict(zip(param_grid.keys(), v)) for v in product(*param_grid.values())]

rmses = []

cutoffs = pd.date_range(start='1963-01-01', end='1969-06-01', freq='6MS')

for param in params:
    m = Prophet(**param)
    m.add_country_holidays(country_name='US')
    m.fit(df)
    
    df_cv = cross_validation(model=m, horizon='365 days', cutoffs=cutoffs)
    df_p = performance_metrics(df_cv, rolling_window=1)
    rmses.append(df_p['rmse'].values[0])
    
tuning_results = pd.DataFrame(params)
tuning_results['rmse'] = rmses

best_params = params[np.argmin(rmses)]

print(best_params)
```

## Fit best model 

```python
m = Prophet(
    changepoint_prior_scale=0.01,
    seasonality_prior_scale=0.01
)
m.add_country_holidays(country_name='US')
m.fit(df);
```

## Make forecasts 

```python
future = m.make_future_dataframe(periods=12, freq='M')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast_fig = m.plot(forecast)
```

```python
components_fig = m.plot_components(forecast)
```

## Performance metrics 

```python
df_cv = cross_validation(model=m, horizon='365 days', cutoffs=cutoffs)
df_p = performance_metrics(df_cv)

fig = plot_cross_validation_metric(df_cv, metric='mape')
```
